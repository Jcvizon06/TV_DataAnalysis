# Objective

The objective of this program is to extract and visualize information of the dataset.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
visual_df = pd.read_csv('Visualization Dataset.csv')
```

```python
visual_df.head(5)
```

```python
visual_df.describe()
```

```python
'''
The Most Expensive TV
'''
print(visual_df.iloc[visual_df['Original Price'].idxmax()])
print('\n')
print(visual_df.iloc[visual_df['Discounted Price'].idxmax()])
```

```python
'''
The Cheapest TV
'''
print(visual_df.iloc[visual_df['Original Price'].idxmin()])
print('\n')
print(visual_df.iloc[visual_df['Discounted Price'].idxmin()])
```

```python
visual_df['Discount Value'] = visual_df['Original Price'] - visual_df['Discounted Price']
```

```python
'''
Highest Discount Value
'''
print(visual_df.iloc[visual_df['Discount Value'].idxmax()])
```

```python
'''
Lowest Discount Value
'''
print(visual_df.iloc[visual_df['Discount Value'].idxmin()])
```

```python
visual_df['Discount Percentage'] = visual_df['Discount Value'] / visual_df['Original Price'] *100
```

```python
'''
Higest Discount Percentage
'''
visual_df.iloc[visual_df['Discount Percentage'].idxmax()]
```

```python
'''
Lowest Discount Percentage
'''
visual_df.iloc[visual_df['Discount Percentage'].idxmin()]
```

```python
'''
Products that are out of stock
'''
visual_df[visual_df['Stock Status']=='Out of stock']
```

```python
'''
Products that has no warranty
'''
visual_df['Product Name'][visual_df['Warranty']==0]
```

```python
'''
Products that has 3 years warranty
'''
visual_df['Product Name'][visual_df['Warranty']==3]
```
